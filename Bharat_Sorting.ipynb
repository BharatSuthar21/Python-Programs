```python
#importing the required module of python
import random
import matplotlib.pyplot as plt
```

```python
################## Variable ##################

# Global variable to store the data 
selection_comparision = 0
selection_swap = 0

insertion_comparision = 0
insertion_swap = 0

quicksort_comparision = 0
quicksort_swap = 0




################lists####################


# lists to store final data of the function to draw the curve
len_list = []
selection_comparision_list = []
selection_swap_list = []

insertion_comparision_list = []
insertion_swap_list = []

quicksort_comparision_list = []
quicksort_swap_list = []
```

```python
##############Insertion Sort###############

def find_index(b,i):

	global insertion_comparision
	j = i
	set_index = i

	while j > 0 and b[j-1] > b[i]:
		set_index = j-1
		insertion_comparision += 1
		j -= 1

	return set_index


def shift_elements(b,n,m):
	global insertion_swap
	temp = b[m]

	for k in range(m,n,-1):
		b[k] = b[k-1]
		insertion_swap +=1

	b[n]=temp
	insertion_swap +=1



def insertion_sort(lst):
	i = 0
	while i < len(lst):
		n = find_index(lst,i)
		shift_elements(lst,n,i)
		i+=1
```

```python
##############Selection Sort################
# function to swap the element of list
def swap (lst,x,y):

	global selection_swap

	temp = lst[x]
	lst[x] = lst[y]
	lst[y] = temp

	selection_swap +=1


# function to find the minimum index 
def findmin(lst, n):

	global selection_comparision

	min_ind = n

	for i in range(n, len(lst)):
		selection_comparision += 1
		if lst[min_ind] > lst[i]:
			min_ind = i

	return min_ind



def selection_sort(lst):
	i = 0
	while i < len(lst):
		a = findmin(lst, i)
		swap(lst, a, i)
		i += 1
```

```python
###########Quick Sort############

def swap_for_quick_sort(lst,x,y):

	global quicksort_swap
	temp = lst[x]
	lst[x] = lst[y]
	lst[y] = temp
	quicksort_swap +=1


def partition(b, h, k):

	global quicksort_comparision

	i = h
	j = k+1
	x = b[h]

	while i < j-1:
		quicksort_comparision +=1
		if b[i+1] >= x:
			swap_for_quick_sort(b,i+1,j-1)
			j = j-1
		else :
			swap_for_quick_sort(b,i,i+1)
			i +=1

	return i


def quick_sort(b, h, k):
	"""
	Sort the array fragment b[h..k]
	"""
	if k-h < 1: #has fewer than 2 elements:
		return b
	j = partition(b, h, k)
	quick_sort(b, h, j-1)
	quick_sort(b, j+1, k)

```

```python
#######################################



for i in range(30,240,10):
	len_list.append(i)
	for j in range(3):
		lst = [random.randint(100,1000) for k in range(i)]
		lst_copy1 = lst
		lst_copy2 = lst


		insertion_sort(lst_copy2)
		selection_sort(lst)
		quick_sort(lst_copy1,0,i-1)



	selection_comparision_list.append(selection_comparision/3)
	selection_swap_list.append(selection_swap/3)
	insertion_comparision_list.append(insertion_comparision/3)
	insertion_swap_list.append(insertion_swap/3)
	quicksort_comparision_list.append(quicksort_comparision/3)
	quicksort_swap_list.append(quicksort_swap/3)



	selection_comparision = 0
	selection_swap = 0
	insertion_comparision = 0
	insertion_swap = 0
	quicksort_comparision = 0
	quicksort_swap = 0
```

```python

plt.plot(len_list, insertion_comparision_list, color = 'red')
plt.plot(len_list, selection_comparision_list, color = 'green')
plt.plot(len_list, quicksort_comparision_list, color = 'blue')

plt.xlabel("Len of list of input taken")
plt.ylabel("Comparision value of diffrent method")

```
